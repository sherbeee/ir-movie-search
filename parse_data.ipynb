```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
df = pd.read_csv('./movie_data.csv',header=0)
display(df)
```

### Tokenizing of plot description

```python
parsed_df = df[['Title','Plot']]

parsed_df['tokens'] = parsed_df.apply(lambda row: word_tokenize(row['Plot']), axis=1)

```

### Removing stop words

```python
stop_words = set(stopwords.words('english'))
for index, row in parsed_df.iterrows():
    word_tokens = row['tokens']
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    row['tokens'] = filtered_sentence
    
display(parsed_df)
```

### Stemming

```python
ps = PorterStemmer()

for index, row in parsed_df.iterrows():
    word_tokens = row['tokens']
    stemmed_sentence = [ps.stem(w) for w in word_tokens]
    row['tokens'] = stemmed_sentence

display(parsed_df)
```

### Exporting to csv

```python
parsed_df.drop(['Plot'], axis=1).to_csv('parsed_data.csv')
```
